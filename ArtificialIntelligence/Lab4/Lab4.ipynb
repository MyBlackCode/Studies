```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([[0,0],   #p1
             [0,1],    #p2
             [1,0],    #p3
             [1,1]])   #p4

d1 = np. array([0,0,0,1])
w1 = np.random.random(3)
w2 = np.random.random(3)
w3 = np.random.random(3)

```

```python
wu1 = 0.1
k1 = 0
m1 = 0
y1 = np.array([])
for j in range(100):
    k1 =0
    m1+=1
    y1=[]
    for i in range(len(x)):
        xx=x[i]
        dd=d1[i]
        
        s = xx[0]* w1[1] + xx[1]*w1[2] + w1[0] * (-1)
        if(s>=0):
            y = 1
            y1 = np.append(y1,y)
        else:
            y = 0
            y1 = np.append(y1,y)
        
        w1[0]= w1[0] + wu1*(dd-y)* (-1)
        w1[1]= w1[1] + wu1*(dd-y)*xx[0]
        w1[2]= w1[2] + wu1*(dd-y)*xx[1]
        
        if d1[i] != y:
                k1+=1
        
    if k1==0:
        print("Nauczono w ",m1," probie")
        break;
        
xx = np.arange(-10,14)
yy1 = -(w1[1]/w1[2]) * xx + (w1[0]/w1[2])
plt.plot(xx, yy1, "b-", label='Prosta perceptornu 1')

bbx = np.array([])
bby = np.array([])
rrx = np.array([])
rry = np.array([])
for i in range(len(x)):
    if x[i][1] > ( -(w1[1]/w1[2]) * x[i][0] + (w1[0]/w1[2])):
        bbx = np.append(bbx,x[i,0])
        bby = np.append(bby,x[i,1])
    else:
        rrx = np.append(rrx,x[i,0])
        rry = np.append(rry,x[i,1])

print("y1=",y1)            
plt.plot(bbx,bby, "bo", label='Punkty y=1')
plt.plot(rrx, rry, "ro", label='Punkty y=0')


plt.axis([-2, 2, -2, 2])
plt.legend()
```
