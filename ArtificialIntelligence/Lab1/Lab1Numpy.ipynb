```python
#Zad1
import numpy as np

macierz = np.random.random([10,10])

print("Najmniejszy element:",np.min(macierz))
print("Największy element:",np.max(macierz))
print()

#Zad2  
print(macierz)
print()
for i in range(0, len(macierz), 1):
    print("Największy element w wierszu",i+1,":",max(macierz[i,:]))
print()
    
for i in range(0, len(macierz), 1):
    print("Największy element w kolumnie",i+1,":",max(macierz[:,i]))
```

```python
#Zad3
import numpy as np

macierz = np.zeros([5,5])
print(macierz)
print()

macierz[:,0] = 1
macierz[:,4] = 1
macierz[0,:] = 1
macierz[4,:] = 1
print(macierz)
print()

def zamienWartosci(m):
    for i in range(0, len(m), 1):
        for j in range(0, len(m), 1):
            if m[i][j]== 1:
                m[i][j] = 0
            else:
                m[i][j] =1
    return m

macierz2 = zamienWartosci(macierz)
print(macierz2)



```

```python
#Zad4
import numpy as np

macierz1 = np.random.random([5,5])
macierz2 = np.random.random([5,5])
print(macierz1)
print()
print(macierz2)
print()

#a
macierz3 = np.add(macierz1, macierz2)
print(macierz3)
print()

#b
macierz4 = np.subtract(macierz1,macierz2)
print(macierz4)
print()

#c
macierz5 = np.dot(macierz1,macierz2)
print(macierz5)
print()

#d
macierz6 = macierz3 *2 /10
print(macierz6)
print()
#e
macierz7 = macierz6[1:4,1:4]
print(macierz7)
print()

#f
wektor = macierz7.flatten()
print(wektor)
```
