```python
#Biblioteka matplotlib wykres 4
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.arange(-5,5,0.25)
y = np.arange(0,10,0.25)
x, y = np.meshgrid(x, y)
z = np.sqrt(np.power(x,2)+np.power(y,2))



surf = ax.plot_surface(x,y,z, cmap=cm.inferno, linewidth=0, antialiased=False)

plt.show()
```

```python
#Generowanie i prezentacja danych 1
import matplotlib.pyplot as plt
import numpy as np
import random
N=100
p = np.random.random([N,2]) * 20 - 10

print(p)
print()
plt.plot(p[:,0], p[:,1],'.',color='black')

x = np.arange(-20,20,0.25)
a=1
b=2
y = a *x +b
for i in range(0,N,1):
    if p[i,1] > 1 * p[i,0] + 2:
        plt.plot(p[i,0], p[i,1],'.',color='grey')
    else:
        plt.plot(p[i,0], p[i,1],'.',color='black')
        
        
plt.plot(x, y, 'r-')
plt.axis([-10, 10, -10, 10])
```

```python
#Generowanie i prezentacja danych 2
import matplotlib.pyplot as plt
import numpy as np
import random
import math
N=100
p = np.random.random([N,2]) * 10 - 5

print(p)
print()
#plt.plot(p[:,0], p[:,1],'.',color='blue')

x = np.arange(-5,5,0.25)
y = 2*np.sin(x) - x/2

d = np.zeros(100)
for i in range(0,N,1):
    if p[i,1] > 2 * math.sin (p[i,0]) - p[i,0]/2:
        plt.plot(p[i,0], p[i,1],'.',color='grey')
        d[i]=1
    else:
        plt.plot(p[i,0], p[i,1],'.',color='black')
        d[i]=0
        
print (d)

plt.plot(x, y, 'r-')
plt.axis([-5, 5, -10, 10])


```

```python
#Wykresy 3D 1
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.arange(-10,10,0.2)
y = np.arange(-6,14,0.2)
x, y = np.meshgrid(x, y)
z = (np.sin(np.power(x,2)+np.power(y,2)))/(np.abs(x*y)+1)

surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
plt.show()
```

```python
#Wykresy 3D 2
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.arange(-10,10,0.2)
y = np.arange(-6,14,0.2)
x, y = np.meshgrid(x, y)
z = np.sqrt(np.power(x,2)+np.power(y,2))+ 3* np.cos(np.sqrt(np.power(x,2)+np.power(y,2)))+5

surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
plt.show()
```
